```python
#Grfico de lineas

import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear un grfico de lneas
plt.plot(x, y, label='Lnea de Ejemplo')

# Personalizar el grfico
plt.title('Grfico de Lneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

# Mostrar el grfico
plt.show()


#GRFICO DE DISPERSIN

import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear un grfico de dispersin
plt.scatter(x, y, label='Puntos de Ejemplo', color='red', marker='o')

# Personalizar el grfico
plt.title('Grfico de Dispersin')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

# Mostrar el grfico
plt.show()


#GRFICOS DE BARRAS

import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [4, 7, 1, 8, 5]

# Crear un grfico de barras
plt.bar(categorias, valores, color='green', alpha=0.7, label='Barras de Ejemplo')

# Personalizar el grfico
plt.title('Grfico de Barras')
plt.xlabel('Categoras')
plt.ylabel('Valores')
plt.legend()

# Mostrar el grfico
plt.show()

#GRFICOS DE BARRAS APILADAS

import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores_1 = [4, 7, 1, 8, 5]
valores_2 = [2, 5, 3, 7, 2]

# Crear un grfico de barras apiladas
plt.bar(categorias, valores_1, color='blue', label='Grupo 1')
plt.bar(categorias, valores_2, bottom=valores_1, color='orange', label='Grupo 2')

# Personalizar el grfico
plt.title('Grfico de Barras Apiladas')
plt.xlabel('Categoras')
plt.ylabel('Valores')
plt.legend()

# Mostrar el grfico
plt.show()


#HISTOGRAMAS

import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
datos = np.random.randn(1000)  # Datos distribuidos normalmente

# Crear un histograma
plt.hist(datos, bins=30, color='purple', alpha=0.7, label='Histograma de Ejemplo')

# Personalizar el grfico
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar el grfico
plt.show()
```

**GRAFICOS AVANZADOS**

```python
#GRAFICOS DE CONTORNO Y SUPERFICIE

import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Grfico de contorno
plt.contour(X, Y, Z, cmap='viridis')
plt.title('Grfico de Contorno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()

# Grfico de superficie
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Grfico de Superficie')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.show()


#BOXPLOTS Y VIOLIN PLOTS

import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Diagrama de caja
plt.boxplot(datos, vert=True, patch_artist=True)
plt.title('Diagrama de Caja')
plt.xticks([1, 2, 3], ['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.show()

# Violin Plot
plt.violinplot(datos, showmedians=True)
plt.title('Violin Plot')
plt.xticks([1, 2, 3], ['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.show()

#GRAFICOS DE RADAR

import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores_grupo1 = [4, 7, 1, 8, 5]
valores_grupo2 = [2, 5, 3, 7, 2]

# Grfico de radar
angles = np.linspace(0, 2 * np.pi, len(categorias), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))  # Repetir el primer ngulo al final para cerrar el grfico

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, valores_grupo1 + [valores_grupo1[0]], 'o-', label='Grupo 1')
ax.fill(angles, valores_grupo1 + [valores_grupo1[0]], alpha=0.25)

ax.plot(angles, valores_grupo2 + [valores_grupo2[0]], 'o-', label='Grupo 2')
ax.fill(angles, valores_grupo2 + [valores_grupo2[0]], alpha=0.25)

# Ajustar las ubicaciones de las etiquetas
ax.set_xticks(angles)

# Ajustar las etiquetas para que coincidan con las ubicaciones
ax.set_xticklabels(categorias + [categorias[0]])

ax.set_title('Grfico de Radar para Comparacin Multivariada')
ax.legend()
plt.show()






```

**GRAFICOS 3D**

```python
#CREACION DE GRAFICO

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Datos de ejemplo
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear una figura y un eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear un grfico de superficie 3D
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Aadir barra de colores
fig.colorbar(surf)

# Etiquetas y ttulo
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Grfico 3D de Superficie')

# Mostrar el grfico
plt.show()


#PERSONALIZACION

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Datos de ejemplo
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(0, 5, 100)
r = z**2 + 1

# Crear una figura y un eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear un grfico de lnea 3D
ax.plot(r * np.cos(theta), r * np.sin(theta), z, label='Ejemplo de Trayectoria 3D')

# Personalizar el grfico
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Grfico 3D de Lnea')
ax.legend()

# Mostrar el grfico
plt.show()

#PALETA DE COLORES

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Datos de ejemplo
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(0, 5, 100)
r = z**2 + 1

# Crear una figura y un eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear un grfico de lnea 3D con paleta de colores
sc = ax.scatter(r * np.cos(theta), r * np.sin(theta), z, c=z, cmap='viridis', label='Ejemplo de Trayectoria 3D')

# Aadir barra de colores
cbar = plt.colorbar(sc)
cbar.set_label('Eje Z')

# Personalizar el grfico
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Grfico 3D de Lnea con Paleta de Colores')
ax.legend()

# Mostrar el grfico
plt.show()


#INSTALACION DE BIBLIOTECA IPYWIDGETS
#pip install ipywidgets
#jupyter nbextension enable --py widgetsnbextension
#jupyter nbextension enable --py widgetsnbextension --sys-prefix

#USO DE SLIDERS Y BOTONES

import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, widgets

# Funcin para graficar una funcin seno con frecuencia ajustable
def plot_sine_wave(frequency):
    x = np.linspace(0, 2 * np.pi, 1000)
    y = np.sin(frequency * x)
    plt.plot(x, y)
    plt.title(f'Funcin Seno con Frecuencia {frequency}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

# Crear un slider interactivo
frequency_slider = widgets.FloatSlider(value=1, min=0.1, max=5.0, step=0.1, description='Frecuencia:')

# Conectar el slider con la funcin usando interact
interact(plot_sine_wave, frequency=frequency_slider)

# Mostrar el widget interactivo
plt.show()


#BOTONES

from ipywidgets import Button

# Funcin para imprimir un mensaje cuando se presiona el botn
def on_button_click(b):
    print("Botn presionado!")

# Crear un botn
button = Button(description="Presionar")

# Conectar la funcin al evento de clic del botn
button.on_click(on_button_click)

# Mostrar el botn
display(button)



```

**SUBPLOTS Y GRFICOS MLTIPLES**

```python
#SUBPLOTS

import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear subgrficos
plt.figure(figsize=(10, 4))

# Subplot 1
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer grfico
plt.plot(x, y1, label='Seno')
plt.title('Grfico de Seno')

# Subplot 2
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo grfico
plt.plot(x, y2, label='Coseno')
plt.title('Grfico de Coseno')

# Ajustar el diseo
plt.tight_layout()

# Mostrar los subgrficos
plt.show()


#SUBPLOT2GRID

import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear subgrficos con subplot2grid para un diseo personalizado
fig = plt.figure(figsize=(10, 4))

# Subplot 1
ax1 = plt.subplot2grid((2, 3), (0, 0), colspan=2)  # 2 filas, 3 columnas, primer grfico, ancho de 2 columnas
ax1.plot(x, y1, label='Seno')
ax1.set_title('Grfico de Seno')

# Subplot 2
ax2 = plt.subplot2grid((2, 3), (0, 2), rowspan=2)  # 2 filas, 3 columnas, tercer grfico, altura de 2 filas
ax2.plot(x, y2, label='Coseno')
ax2.set_title('Grfico de Coseno')

# Subplot 3
ax3 = plt.subplot2grid((2, 3), (1, 0), colspan=2)  # 2 filas, 3 columnas, cuarto grfico, ancho de 2 columnas
ax3.plot(x, y1 + y2, label='Seno + Coseno', color='orange')
ax3.set_title('Suma de Seno y Coseno')

# Ajustar el diseo
fig.tight_layout()

# Mostrar los subgrficos
plt.show()

```

**COLORMAPS**

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un grfico con un colormap personalizado
plt.figure(figsize=(8, 4))
sc = plt.scatter(x, y, c=y, cmap='coolwarm', s=50, edgecolors='k')  # 'coolwarm' es el colormap

# Personalizar la barra de colores
cbar = plt.colorbar(sc)
cbar.set_label('Valor de Y')

# Personalizar el grfico
plt.title('Grfico con Colormap Personalizado')
plt.xlabel('X')
plt.ylabel('Y')

# Mostrar el grfico
plt.show()
```
